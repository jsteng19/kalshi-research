```python
import glob

# Get all transcript files
transcript_files = glob.glob("data-harris/processed-transcripts/*/*.txt")

# Check each file and delete if empty
for file in transcript_files:
    with open(file, 'r', encoding='utf-8') as f:
        content = f.read().strip()
        if not content:  # If file is empty or only whitespace
            os.remove(file)
            print(f"Deleted empty file: {file}")

print(f"Total Harris transcript files: {len(glob.glob('data-harris/processed-transcripts/*/*.txt'))}")
```
